```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("BlackFriday.csv")
```

```python
del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
sns.countplot(x = df['Stay_In_Current_City_Years'])
```

```python
df.groupby('Stay_In_Current_City_Years').size().plot(kind = 'pie' , autopct = "%0.1f%%")
```

```python
sns.countplot(x = 'Stay_In_Current_City_Years' , hue = 'Gender' , data = df)
```

```python
sns.countplot(x = 'Stay_In_Current_City_Years' , hue = 'Marital_Status' , data = df)
```

```python

print(df['Stay_In_Current_City_Years'].dtype)
print(df['Marital_Status'].dtype)
```

```python
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].astype(str)
df['Marital_Status'] = df['Marital_Status'].astype(str)
```

```python
sns.countplot(x = 'Stay_In_Current_City_Years', hue = "Marital_Status" , data = df)
```
